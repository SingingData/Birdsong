```python
count = 0

primary_input = []
native_wav_data = []
for samples in train_files:
    count = count + 1
    os.chdir('..')
    os.chdir('./resampled-train')
    [fs, s] = wavfile.read(samples) 
    s = np.asarray(s[0:228800])
    native_wav_data.append(s)    
    os.chdir('..')
    os.chdir('./raw-train-arrays')
    np.savetxt('train_raw_data'+str(count)+'.csv', s, delimiter=',')
native_wav_data = np.array(native_wav_data)
print(native_wav_data.shape)
```

```python
count = 0

primary_input = []
native_wav_data = []
for samples in test_files:
    count = count + 1
    os.chdir('..')
    os.chdir('./resampled-test')
    [fs, s] = wavfile.read(samples) 
    s = np.asarray(s[0:228800])
    native_wav_data.append(s)    
    os.chdir('..')
    os.chdir('./raw-test-arrays')
    np.savetxt('test_raw_data'+str(count)+'.csv', s, delimiter=',')
native_wav_data = np.array(native_wav_data)
print(native_wav_data.shape)
```
